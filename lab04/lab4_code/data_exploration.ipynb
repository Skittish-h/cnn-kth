```python
from util import read_mnist
import numpy as np
import matplotlib.pyplot as plt

# Load the data
train_images, train_labels, test_images, test_labels = read_mnist(n_train=60000, n_test=10000)

print(f"Train images shape: {train_images.shape}")
print(f"Train labels shape: {train_labels.shape}")
print(f"Test images shape: {test_images.shape}")
print(f"Test labels shape: {test_labels.shape}")
```

```python
def plot_sample_images(images, labels, n_samples=5):
    plt.figure(figsize=(10, 2))
    for i in range(n_samples):
        plt.subplot(1, n_samples, i + 1)
        plt.imshow(images[i].reshape(28, 28), cmap='gray')
        plt.title(f"Label: {np.argmax(labels[i])}")
        plt.axis('off')
    plt.show()

plot_sample_images(train_images, train_labels, n_samples=5)
```

```python
def plot_label_distribution(labels, title="Label Distribution"):
    label_counts = np.sum(labels, axis=0)
    plt.bar(range(10), label_counts)
    plt.xlabel("Digit Label")
    plt.ylabel("Count")
    plt.title(title)
    plt.show()

plot_label_distribution(train_labels, title="Training Label Distribution")
plot_label_distribution(test_labels, title="Test Label Distribution")
```
